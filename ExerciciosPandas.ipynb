1) Utilizando o DataFrame importado (alunos3.csv) calcule a média das provas em uma nova coluna chamada (Media_provas)

```python
notas_provas = pd.read_csv('dados/alunos3.csv', sep=';')
notas_provas['Prova_1'] = notas_provas['Prova_1'].str.replace(',', '.')
notas_provas['Prova_2'] = notas_provas['Prova_2'].str.replace(',', '.')
notas_provas['Prova_3'] = notas_provas['Prova_3'].str.replace(',', '.')
notas_provas
```

```python
notas_provas.info()
```

```python
notas_provas['Prova_1'] = notas_provas['Prova_1'].astype(float)
notas_provas['Prova_2'] = notas_provas['Prova_2'].astype(float)
notas_provas['Prova_3'] = notas_provas['Prova_3'].astype(float)
notas_provas['Prova_4'] = notas_provas['Prova_4'].astype(float)
```

```python
notas_provas['Media_provas'] = ((notas_provas['Prova_1'] + notas_provas['Prova_2'] + notas_provas['Prova_3'] + notas_provas['Prova_4']) / 4)
notas_provas
```

2) Quem foram os alunos que obtiveram a maior e a menor média

```python
maior_nota = notas_provas['Media_provas'].max()
maior_nota_aluno = notas_provas[notas_provas['Media_provas'] == maior_nota]
maior_nota_aluno
```

```python
menor_nota = notas_provas['Media_provas'].min()
menor_nota_aluno = notas_provas[notas_provas['Media_provas'] == menor_nota]
menor_nota_aluno
```

3) Agora una este dataframe com o `cadastro_alunos.xlsx`

```python
cadastro_alunos = pd.read_excel('dados/cadastro_alunos.xlsx')
cadastro_alunos
```

```python
dados_alunos = cadastro_alunos.merge(notas_provas, how='left', left_on='RA', right_on='RA')
dados_alunos
```

4) Qual a média entre as Media_provas dentro do público feminino? e masculino?

```python
media_masc = dados_alunos['Media_provas'].where(dados_alunos['Sexo'] == 'Masculino').mean()
media_masc
```

5) Qual a média de idade das pessoas que obtiveram Media_provas maior ou igual a 7?

```python
media_idade = dados_alunos['Idade'].where(dados_alunos['Media_provas'] >= 7).mean()
media_idade
```

6) Qual das cidades possui o maior média de Media_provas? E qual é este valor?

```python
media_cidade = dados_alunos.groupby('Cidade').agg({'Media_provas': ['mean']})

print(media_cidade)
maior_media_cidade = media_cidade.idxmax()
print(maior_media_cidade)
print(media_cidade.max())
```
